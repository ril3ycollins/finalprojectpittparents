Our project is an analysis where we found the best neighborhood in Pittsburgh to raise children according to which neighborhood has the highest number of playgrounds, courts, and pools. Our metric is raisability of children. We used how many pools, playgrounds, and courts/rinks were in each neighborhood to find the best possible neighborhood for a child to grow up and enjoy playing in.
When finding our datasets we made sure that all three had a "neighborhood" category so they could be related and easily combined. Some datasets only had zip codes, some didn't, which made it harder to find three sets that could fit together.


```python
import pandas as pd
import matplotlib.pyplot as plt
playgroundData = pd.read_csv("https://data.wprdc.org/datastore/dump/47350364-44a8-4d15-b6e0-5f79ddff9367")

playgroundsPerNeighborhood = playgroundData['neighborhood'].value_counts()
playgroundsPerNeighborhood = playgroundsPerNeighborhood[playgroundsPerNeighborhood > 1]

playgroundsPerNeighborhood.plot.bar(color="pink")
plt.title("Number of Playgrounds per Neighborhood in Pittsburgh (With Over One Instance)")
plt.xlabel("Neighborhood")
plt.ylabel("Instances")
plt.xticks(fontsize=9)
plt.tight_layout()
plt.show()

courts_data = pd.read_csv('https://data.wprdc.org/datastore/dump/a5b71bfa-840c-4c86-8f43-07a9ae854227')
df = courts_data['neighborhood'].value_counts()
df20 = df.head(30)
df20.plot.bar(color={"turquoise"})
plt.tight_layout()
plt.title("Top Pittsburgh Neighboorhoods by Number of Courts/Rinks")
plt.show()

pool_data = pd.read_csv('https://data.wprdc.org/datastore/dump/5cc254fe-2cbd-4912-9f44-2f95f0beea9a')
df2 = pool_data['neighborhood'].value_counts()
df2.plot.bar(color = {"slateblue"})
plt.title ("Number of Pools per Neighborhood in Pittsburgh")
plt.show()

playgroundsData = pd.read_csv("https://data.wprdc.org/datastore/dump/47350364-44a8-4d15-b6e0-5f79ddff9367")
courtsData = pd.read_csv('https://data.wprdc.org/datastore/dump/a5b71bfa-840c-4c86-8f43-07a9ae854227')
poolData = pd.read_csv('https://data.wprdc.org/datastore/dump/5cc254fe-2cbd-4912-9f44-2f95f0beea9a')
playgrounds_per_neighborhood = playgroundsData['neighborhood'].value_counts()
courts_per_neighborhood = courtsData['neighborhood'].value_counts()
pools_per_neighborhood = poolData['neighborhood'].value_counts()

combined_data = pd.DataFrame({
    'Playgrounds': playgrounds_per_neighborhood,
    'Courts/Rinks': courts_per_neighborhood,
    'Pools': pools_per_neighborhood
})

combined_data['Total'] = combined_data.sum(axis=1)
top_neighborhoods = combined_data.sort_values(by='Total', ascending=False).head(10)
top_neighborhoods = top_neighborhoods.drop(columns='Total')

top_neighborhoods.plot(kind='barh', color=['pink', 'turquoise', 'slateblue'])
plt.title("Top 10 Neighborhoods in Pittsburgh by Number of Playgrounds, Courts/Rinks, and Pools in each")
plt.xlabel("Count")
plt.ylabel("Neighborhood")
plt.yticks(fontsize=6)
plt.legend(title="Type")
plt.show()
```

This graph combines the previous 3 graphs. It shows the top 10 neighborhoods in Pittsburgh with the most playgrounds, courts/rinks, and pools. In conclusion, Squirrel Hill South is the overall best neighborhood for kids to live in Pittsburgh. While it has one less pool than second place, Highland Park, the number of courts/rinks and playgrounds in Squirrel Hill South outweighs the extra pool in Highland Park.
