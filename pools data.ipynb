# Pool Data Final Project

In order to find the best neighborhood, my group took the perspective of a kid. Because of this, I chose a dataset recording the pools and spray parks in Pittsburgh neighborhoods and their locations. The best neighborhoods are the ones pools and/or sprayparks. These features add to the childhood funness of the area and are what make life worth living!

```python
import pandas as pd
pool_data = pd.read_csv('https://data.wprdc.org/datastore/dump/5cc254fe-2cbd-4912-9f44-2f95f0beea9a')
```

```python
df = pool_data['neighborhood'].value_counts()
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df.plot.bar(color = {"slateblue"})
plt.show()
```

In conclusion, South Side Flats and Highland Park and the very best Pittsburgh neighborhoods due to the abundance of pools and/or spray parks (2 total).The remaining neighborhoods are tied for second place with 1 pool.
