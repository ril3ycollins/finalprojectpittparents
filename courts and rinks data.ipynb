```python
import pandas as pd
import matplotlib.pyplot as plt

courts_data = pd.read_csv('https://data.wprdc.org/datastore/dump/a5b71bfa-840c-4c86-8f43-07a9ae854227')

types = courts_data['type'].unique()
typesChart = courts_data['type'].value_counts()
typesChart.plot(kind='pie', autopct='%1.1f%%')
plt.tight_layout()
plt.ylabel('')
plt.show()

print("This pie chart shows the different kinds of courts and rinks that can be found across all Pittsburgh neighborhoods. The types are: ", end="")
for x in types:
    if x == "Horseshoe Pit":
        print("and "+x+".")
    else:
        print(x, end=", ")

print("\n")


df = courts_data['neighborhood'].value_counts()
df20 = df.head(30)
df20.plot.bar(color={"turquoise"})
plt.tight_layout()
plt.title("Top Pittsburgh Neighboorhoods by Number of Courts/Rinks")
plt.show()

y = """This graph shows the top 30 neighborhoods in Pittsburgh based on the number of courts and rinks
in the neighborhood. This metric is very important for raising a child so they have the oppertunity to
be well versed in many sports such as tennis, hockey, basketball, and more. The neighborhood with the
highest number of courts/rinks in Pittsburgh is Squirrel Hill South with 26 courts/rinks. Highland Park 
is a close second with 20 courts/rinks."""

print(y)
```
