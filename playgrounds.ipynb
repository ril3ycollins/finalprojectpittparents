```python
import pandas as pd
import matplotlib.pyplot as plt

playgroundData = pd.read_csv("https://data.wprdc.org/datastore/dump/47350364-44a8-4d15-b6e0-5f79ddff9367")
```

```python
playgroundsPerNeighborhood = playgroundData['neighborhood'].value_counts()
playgroundsPerNeighborhood = playgroundsPerNeighborhood[playgroundsPerNeighborhood > 1]
print(playgroundData['neighborhood'].value_counts())
print("")
print("There are 68 Neighborhoods in the dataset.")
```

```python

playgroundsPerNeighborhood.plot.bar(color="pink")
plt.title("Number of Playgrounds per Neighborhood in Pittsburgh (With Over One Instance)")
plt.xlabel("Neighborhood")
plt.ylabel("Instances")
plt.xticks(fontsize=9)
plt.tight_layout()
plt.show()
```

```python
print("This graph shows the number of playgrounds per neighborhood in Pittsburgh, not including the neighborhoods with only one playground for readabilitys sake.")
print("The best neighborhood according to my playground metric is Squirrel Hill South, coming in first place with 8 playgrounds. The two ties for second place, Beechview and South Side Slopes, both only have 5 playgrounds, which showcases how Squirrel Hill South is much more superior.")
print("This data is important because children should have a variety of places to have fun at and interact with other children. The more playgrounds, the more child friendly the neighborhood is.")
```
